```python
#打印要素图层的特性类
import arcpy
element = r"E:\code\gis\pro\pymap\pymap.gdb\lines"
desc = arcpy.da.Describe(element)
print("Dataset type: " + desc["datasetType"])
```

```python
#通过da.Describe获取要素的可以访问许多其他相关属性。这些属性包括文件路径、目录路径、名称、文件名和基本名称
import arcpy
arcpy.env.workspace = r"E:\code\gis\pro\pymap\pymap.gdb"
element = "rout_Buffer"
desc = arcpy.da.Describe(element)
print("Data type: " + desc["dataType"])
print("File path: " + desc["path"])
print("Catalog path: " + desc["catalogPath"])
print("File name: " + desc["file"])
print("Base name: " + desc["baseName"])
print("Name: " + desc["name"])
```
